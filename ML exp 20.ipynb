```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample Dataset
# Months
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)

# Sales
y = np.array([100, 120, 140, 160, 180, 200])

# Create Model
model = LinearRegression()

# Train Model
model.fit(X, y)

# Predict Future Sales for Month 7
future_sales = model.predict([[7]])

print("Predicted Sales:", future_sales[0])
```
