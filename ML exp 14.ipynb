```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample Dataset
# Feature: House Area (sq.ft)
X = np.array([[1000], [1200], [1500], [1800], [2000]])

# House Prices
y = np.array([3000000, 3600000, 4500000, 5400000, 6000000])

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict price for a 1600 sq.ft house
predicted_price = model.predict([[1600]])

print("Predicted House Price:", predicted_price[0])
```
