```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample Dataset
# Features: [Car Age (years), Mileage (km)]
X = np.array([
    [1, 10000],
    [2, 20000],
    [3, 30000],
    [4, 40000],
    [5, 50000]
])

# Car Prices
y = np.array([900000, 800000, 700000, 600000, 500000])

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict price of a car
# Age = 3 years, Mileage = 35000 km
predicted_price = model.predict([[3, 35000]])

print("Predicted Car Price:", predicted_price[0])
```
