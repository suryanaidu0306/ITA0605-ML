```python
from sklearn.tree import DecisionTreeClassifier

# Sample Dataset
# Features: [RAM(GB), Storage(GB)]
X = [
    [2, 32],
    [3, 64],
    [4, 64],
    [6, 128],
    [8, 128],
    [12, 256]
]

# Price Categories
# 0 = Low, 1 = Medium, 2 = High
y = [0, 0, 1, 1, 2, 2]

# Create model
model = DecisionTreeClassifier()

# Train model
model.fit(X, y)

# Predict mobile price category
sample = [[8, 256]]
prediction = model.predict(sample)

print("Predicted Price Category:", prediction[0])
```
