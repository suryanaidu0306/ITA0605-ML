```python
from sklearn.naive_bayes import GaussianNB

# Sample Dataset
# Features: [Income, Credit Score]
X = [
    [25000, 600],
    [30000, 650],
    [40000, 700],
    [50000, 750],
    [60000, 800],
    [70000, 850]
]

# Loan Status
# 0 = Rejected, 1 = Approved
y = [0, 0, 0, 1, 1, 1]

# Create Naive Bayes Model
model = GaussianNB()

# Train Model
model.fit(X, y)

# Predict Loan Status
sample = [[55000, 780]]
prediction = model.predict(sample)

print("Loan Status:", prediction[0])
```
